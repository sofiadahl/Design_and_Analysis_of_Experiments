```python


#If you have same columns in all your csv files 
#header=0 so that after reading csv first row can be assigned as the column names.
# Nstrokes; Tprep (Mean/SD); Zprep (Mean/SD); Tstroke (Mean/SD); MAGstroke (Mean/SD); MAGreb (Mean/SD), Ratio (Mean/SD)

import pandas as pd
import glob
import statistics as stat
import matplotlib.pyplot as plt

header=(['player','ref','pattern', 'style','Nstrokes','Tprep_mean','Tprep_SD','Zprep_mean','Zprep_SD','Tstroke_mean','Tstroke_SD','MAGstroke_mean','MAGstroke_SD','MAGreb_mean','MAGreb_SD','Ratio_mean','Ratio_SD'])
len(header)
```

```python

path = r'/Users/sofia/Documents/PROJECTS/TIME-UiO/Mocapanalysis/2020Analysis/May2020/meanfeatures/' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
   # df = pd.read_csv(filename, index_col=None, header=0)
    df = pd.read_csv(filename, index_col=None, header=None, names=header)
    #    df = pd.read_csv(filename, index_col=None, header=None)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame.describe()
```
